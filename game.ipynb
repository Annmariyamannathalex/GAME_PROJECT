```python
class HealingSkill:
    def __init__(self, name, heal_points, skill_type):
        self.name = name
        self.heal_points = heal_points
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Heal Points: {self.heal_points}, Type: {self.skill_type}")

    def use_skill(self, target):
        print(f"Player uses {self.name}!")
        target.hp += self.heal_points
        print(f"{self.name} restores {self.heal_points} HP.")


class RegenerationSkill:
    def __init__(self, name, heal_per_turn, skill_type):
        self.name = name
        self.heal_per_turn = heal_per_turn
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Heal per Turn: {self.heal_per_turn}, Type: {self.skill_type}")

    def use_skill(self, target):
        print(f"Player uses {self.name}!")
        target.hp += self.heal_per_turn
        print(f"{self.name} restores {self.heal_per_turn} HP.")


class PowerStrikeSkill:
    def __init__(self, name, damage, skill_type):
        self.name = name
        self.damage = damage
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Damage: {self.damage}, Type: {self.skill_type}")

    def use_skill(self, target):
        print(f"Player uses {self.name}!")
        target.hp -= self.damage
        print(f"{self.name} deals {self.damage} damage to the enemy.")


class PlayerCharacter:
    def __init__(self, hp, max_hp, ap, dp, sp, mp, max_mp, image, job_class, magic_skills):
        self.hp = hp
        self.max_hp = max_hp
        self.ap = ap
        self.dp = dp
        self.sp = sp
        self.mp = mp
        self.max_mp = max_mp
        self.image = image
        self.job_class = job_class
        self.magic_skills = magic_skills

    def reset_stats(self):
        self.hp = self.max_hp
        self.mp = self.max_mp


# Create instances of the HealingSkill, RegenerationSkill, and PowerStrikeSkill classes
healing_skill = HealingSkill("Healing Light", 20, "Heal")
regeneration_skill = RegenerationSkill("Regeneration", 10, "Heal Over Time")
power_strike_skill = PowerStrikeSkill("Power Strike", 30, "Damage")

# Create a warrior player character
warrior = PlayerCharacter(
    hp=100,
    max_hp=100,
    ap=20,
    dp=15,
    sp=10,
    mp=5,
    max_mp=5,
    image="A strong and heavily-armed warrior",
    job_class="Warrior",
    magic_skills=[healing_skill, regeneration_skill, power_strike_skill]
)

enemy_hp = 100
enemy_ap = 10
action = 0
#Battle
player

while True:
    print("Player action:")
    print("1) Attack")
    print("2) Quit")
    print("3) Restart")
    print("4) Use Skill")
    action = int(input("Enter your action: "))

    if action == 1:
        print("Player attacking")
        enemy_hp -= warrior.ap
        print("Enemy Health:", enemy_hp)
        if enemy_hp <= 0:
            print("Player wins!")
            break
    elif action == 2:
        print("Player Quits")
        break
    elif action == 3:
        print("Player Restart")
        warrior.reset_stats()
        enemy_hp = 100
        enemy_ap = 10
        continue
    elif action == 4:
        print("Select a Skill:")
        for i, skill in enumerate(warrior.magic_skills):
            print(f"{i+1}) {skill.name}")
        skill_choice = int(input("Enter the skill number: "))

        if 1 <= skill_choice <= len(warrior.magic_skills):
            skill = warrior.magic_skills[skill_choice - 1]
            skill.use_skill(warrior)
        else:
            print("Invalid skill choice!")

    print("Enemy attacking")
    warrior.hp -= enemy_ap
    print("Player Health:", warrior.hp)
    if warrior.hp <= 0:
        print("Player Loses!")
        break

        
```
