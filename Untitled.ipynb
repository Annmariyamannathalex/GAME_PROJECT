```python
import random

class HealSkill:
    def __init__(self, name, power, skill_type, mp_cost):
        self.name = name
        self.power = power
        self.skill_type = skill_type
        self.mp_cost = mp_cost

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, user):
        if user.mp >= self.mp_cost:
            user.mp -= self.mp_cost
            if(user.hp + self.power) > user.max_hp:
                user.hp = user.max_hp
            else:
                user.hp += self.power
            print(f"{user.job_class} uses {self.name}.")
            print(f"{user.job_class}'s health increased by {self.power}. New HP: {user.hp}")
            print(f"{user.job_class}'s MP reduced by {self.mp_cost}. New MP: {user.mp}")
        else:
            print("Not enough MP to use this skill.")
            
class BackstabSkill:
    def __init__(self, name, power, skill_type):
        self.name = name
        self.power = power
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, user, target):
        damage = self.power - target.dp
        target.hp -= damage
        print(f"{user.job_class} uses {self.name} on {target.enemy_type}.")
        print(f"{target.enemy_type} receives {damage} damage. New HP: {target.hp}")


class ShadowstepSkill:
    def __init__(self, name, power, skill_type):
        self.name = name
        self.power = power
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, user, target):
        damage = self.power - target.dp
        target.hp -= damage
        print(f"{user.job_class} uses {self.name} on {target.enemy_type}.")
        print(f"{target.enemy_type} receives {damage} damage. New HP: {target.hp}")






class VitalityBoostSkill:
    def __init__(self, name, power, skill_type, mp_cost):
        self.name = name
        self.power = power
        self.skill_type = skill_type
        self.mp_cost = mp_cost

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, user):
        if user.mp >= self.mp_cost:
            user.mp -= self.mp_cost
            user.hp += self.power
            print(f"{user.job_class} uses {self.name}.")
            print(f"{user.job_class}'s health increased by {self.power}. New HP: {user.hp}")
            print(f"{user.job_class}'s MP reduced by {self.mp_cost}. New MP: {user.mp}")
        else:
            print("Not enough MP to use this skill.")


class PlayerCharacter:
    def __init__(self, hp, max_hp, ap, dp, sp, mp, max_mp, image, job_class, magic_skills):
        self.hp = hp
        self.max_hp = max_hp
        self.ap = ap
        self.dp = dp
        self.sp = sp
        self.mp = mp
        self.max_mp = max_mp
        self.image = image
        self.job_class = job_class
        self.magic_skills = magic_skills
        self.dp_active = False

    def reset_stats(self):
        self.hp = self.max_hp
        self.mp = self.max_mp

    def defend(self):
        if self.dp_active == True:
            self.dp_active =False
            self.dp -=5
            print(f"{self.job_class} defence back to normal. Defense decreased by 5.")
        else:
            self.dp += 5
            self.dp_active = True
            print(f"{self.job_class} is defending. Defense increased by 5.")
       
    

    def get_speed_points(self):
        return self.sp


class EnemyCharacter:
    def __init__(self, hp, ap, dp, sp, image, enemy_type):
        self.hp = hp
        self.max_hp = hp
        self.ap = ap
        self.dp = dp
        self.sp = sp
        self.image = image
        self.enemy_type = enemy_type

    def get_speed_points(self):
        return self.sp


# Create player characters
warrior = PlayerCharacter(
    hp=100,
    max_hp=100,
    ap=20,
    dp=15,
    sp=10,
    mp=5,
    max_mp=5,
    image="A strong and heavily-armed warrior",
    job_class="Warrior",
    magic_skills=[
        HealSkill("Healing Light", 20, "Heal", 2),
        BackstabSkill("Backstab", 40, "Physical")
    ]
)

rogue = PlayerCharacter(
    hp=100,
    max_hp=100,
    ap=25,
    dp=10,
    sp=15,
    mp=15,
    max_mp=15,
    image="A nimble and stealthy rogue",
    job_class="Rogue",
    magic_skills=[
        ShadowstepSkill("Shadowstep", 35, "Physical"),
        VitalityBoostSkill("Vitality Boost", 30, "Buff", 3)
    ]
)

# Create enemy characters
goblin = EnemyCharacter(
    hp=100,
    ap=10,
    dp=5,
    sp=6,
    image="A small and dangerous creature",
    enemy_type="Goblin"
)

orc = EnemyCharacter(
    hp=100,
    ap=15,
    dp=4,
    sp=6,
    image="A brutal and aggressive character",
    enemy_type="Orc"
)

def get_turn_order(player, enemy):
    player_sp = player.get_speed_points()
    enemy_sp = enemy.get_speed_points()

    if player_sp > enemy_sp:
        return player, enemy
    elif player_sp < enemy_sp:
        return enemy, player
    else:
        return player, enemy if isinstance(player, PlayerCharacter) else enemy, player


while True:
    print("Select a Player Character:")
    print("1) Warrior")
    print("2) Rogue")
    player_choice = int(input("Enter player number: "))

    if player_choice == 1:
        player = warrior
        print("Player Character: Warrior")
    elif player_choice == 2:
        player = rogue
        print("Player Character: Rogue")
    else:
        print("Invalid player choice")
        continue

    # Select a random enemy character
    enemy = random.choice([goblin, orc])
    print("Enemy type:", enemy.enemy_type)

    # Display player action menu
    print("Player action:")
    print("1) Attack")
    print("2) Defend")
    print("3) Quit")
    print("4) Restart")
    print("5) Use Skill")

    # Get input from player
    action = int(input("Enter action number: "))

    # Process player action
    if action == 1:
        # Player attacks enemy
        print(f"{player.job_class} attacking")
        enemy.hp -= (player.ap - enemy.dp)
        
        print("Enemy Health:", enemy.hp)
        if enemy.hp <= 0:
            print("Player wins!")
            break
    elif action == 2:
        # Player defends
        player.defend()
    elif action == 3:
        print("Player Quits")
        break
    elif action == 4:
        print("Player Restarts")
        player.reset_stats()
        for enemy in [goblin, orc]:
            enemy.hp = enemy.max_hp
        continue
    elif action == 5:
        # Player uses skill
        print(f"Select a Skill for {player.job_class}:")
        for index, skill in enumerate(player.magic_skills, start=1):
            print(f"{index} {skill.name}")
        skill_choice = int(input("Enter skill number: "))
        
        if skill_choice >= 1 and skill_choice <= len(player.magic_skills):
            skill = player.magic_skills[skill_choice - 1]
        if isinstance(skill, VitalityBoostSkill):
            skill.use_skill(player)
        elif isinstance(skill, HealSkill):
            skill.use_skill(player)
        else:
            if hasattr(skill, 'mp_cost'):
                if skill.mp_cost > player.mp:
                    print("Not enough MP to use this skill.")
                else:
                    skill.use_skill(player, enemy)
            else:
                skill.use_skill(player, enemy)
    else:
        print("Invalid skill choice")


        
        print(f"{player.job_class} Health:", player.hp)
        print("Enemy Health:", enemy.hp)
        
    # Enemy attacks player
    print(f"{enemy.enemy_type} attacking")
    player.hp -= enemy.ap
    if player.dp_active == True :
        player.defend()
    print(f"{player.job_class} Health:", player.hp)
    if player.hp <= 0:
        print("player loses!")
        break

    # Check win/lose conditions
    if player.hp <= 0:
        print("Player loses!")
        break
    elif enemy.hp <= 0:
        print("Player wins!")
        break
```
