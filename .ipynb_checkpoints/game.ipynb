```python
import random

class HealSkill:
    def __init__(self, name, power, skill_type, mp_cost):
        self.name = name
        self.power = power
        self.skill_type = skill_type
        self.mp_cost = mp_cost

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, user, target):
        if user.mp >= self.mp_cost:
            user.mp -= self.mp_cost
            target.hp += self.power
            print(f"{user.job_class} uses {self.name} on {target.enemy_type}.")
            print(f"{player.job_class}'s health increased by {self.power}. New HP: {target.hp}")
            print(f"{user.job_class}'s MP reduced by {self.mp_cost}. New MP: {user.mp}")
        else:
            print("Not enough MP to use this skill.")


class BackstabSkill:
    def __init__(self, name, power, skill_type):
        self.name = name
        self.power = power
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, user, target):
        damage = self.power - target.dp
        target.hp -= damage
        print(f"{user.job_class} uses {self.name} on {target.enemy_type}.")
        print(f"{target.enemy_type} receives {damage} damage. New HP: {target.hp}")


class ShadowstepSkill:
    def __init__(self, name, power, skill_type):
        self.name = name
        self.power = power
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, user, target):
        damage = self.power - target.dp
        target.hp -= damage
        print(f"{user.job_class} uses {self.name} on {target.enemy_type}.")
        print(f"{target.enemy_type} receives {damage} damage. New HP: {target.hp}")


class VitalityBoostSkill:
    def __init__(self, name, power, skill_type, mp_cost):
        self.name = name
        self.power = power
        self.skill_type = skill_type
        self.mp_cost = mp_cost

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, user):
        if user.mp >= self.mp_cost:
            user.mp -= self.mp_cost
            user.hp += self.power
            print(f"{user.job_class} uses {self.name}.")
            print(f"{user.job_class}'s health increased by {self.power}. New HP: {user.hp}")
            print(f"{user.job_class}'s MP reduced by {self.mp_cost}. New MP: {user.mp}")
        else:
            print("Not enough MP to use this skill.")


class PlayerCharacter:
    def __init__(self, hp, max_hp, ap, dp, sp, mp, max_mp, image, job_class, magic_skills):
        self.hp = hp
        self.max_hp = max_hp
        self.ap = ap
        self.dp = dp
        self.sp = sp
        self.mp = mp
        self.max_mp = max_mp
        self.image = image
        self.job_class = job_class
        self.magic_skills = magic_skills

    def reset_stats(self):
        self.hp = self.max_hp
        self.mp = self.max_mp

    def defend(self):
        self.dp += 5
        print(f"{self.job_class} is defending. Defense increased by 5.")

    def get_speed_points(self):
        return self.sp


class EnemyCharacter:
    def __init__(self, hp, ap, dp, sp, image, enemy_type):
        self.hp = hp
        self.max_hp = hp
        self.ap = ap
        self.dp = dp
        self.sp = sp
        self.image = image
        self.enemy_type = enemy_type

    def get_speed_points(self):
        return self.sp


# Create player characters
warrior = PlayerCharacter(
    hp=100,
    max_hp=100,
    ap=20,
    dp=15,
    sp=10,
    mp=5,
    max_mp=5,
    image="A strong and heavily-armed warrior",
    job_class="Warrior",
    magic_skills=[
        HealSkill("Healing Light", 20, "Heal", 2),
        BackstabSkill("Backstab", 40, "Physical")
    ]
)

rogue = PlayerCharacter(
    hp=100,
    max_hp=100,
    ap=25,
    dp=10,
    sp=15,
    mp=15,
    max_mp=15,
    image="A nimble and stealthy rogue",
    job_class="Rogue",
    magic_skills=[
        ShadowstepSkill("Shadowstep", 35, "Physical"),
        VitalityBoostSkill("Vitality Boost", 30, "Buff", 3)
    ]
)

# Create enemy characters
goblin = EnemyCharacter(
    hp=100,
    ap=10,
    dp=5,
    sp=6,
    image="A small and dangerous creature",
    enemy_type="Goblin"
)

orc = EnemyCharacter(
    hp=100,
    ap=15,
    dp=4,
    sp=6,
    image="A brutal and aggressive character",
    enemy_type="Orc"
)

def get_turn_order(player, enemy):
    player_sp = player.get_speed_points()
    enemy_sp = enemy.get_speed_points()

    if player_sp > enemy_sp:
        return player, enemy
    elif player_sp < enemy_sp:
        return enemy, player
    else:
        return player, enemy if isinstance(player, PlayerCharacter) else enemy, player


while True:
    print("Select a Player Character:")
    print("1) Warrior")
    print("2) Rogue")
    player_choice = int(input("Enter player number: "))

    if player_choice == 1:
        player = warrior
        print("Player Character: Warrior")
    elif player_choice == 2:
        player = rogue
        print("Player Character: Rogue")
    else:
        print("Invalid player choice")
        continue

    # Select a random enemy character
    enemy = random.choice([goblin, orc])
    print("Enemy type:", enemy.enemy_type)

    # Display player action menu
    print("Player action:")
    print("1) Attack")
    print("2) Defend")
    print("3) Quit")
    print("4) Restart")
    print("5) Use Skill")

    # Get input from player
    action = int(input("Enter action number: "))

    # Process player action
    if action == 1:
        # Player attacks enemy
        print(f"{player.job_class} attacking")
        enemy.hp -= player.ap - enemy.dp
        print("Enemy Health:", enemy.hp)
        if enemy.hp <= 0:
            print("Player wins!")
            break
    elif action == 2:
        # Player defends
        player.defend()
    elif action == 3:
        print("Player Quits")
        break
    elif action == 4:
        print("Player Restarts")
        player.reset_stats()
        for enemy in [goblin, orc]:
            enemy.hp = enemy.max_hp
        continue
    elif action == 5:
        # Player uses skill
        print(f"Select a Skill for {player.job_class}:")
        for index, skill in enumerate(player.magic_skills, start=1):
            print(f"{index}) {skill.name}")
        skill_choice = int(input("Enter skill number: "))

        if skill_choice >= 1 and skill_choice <= len(player.magic_skills):
            skill = player.magic_skills[skill_choice - 1]
            if isinstance(skill, VitalityBoostSkill):
                skill.use_skill(player)
            else:
                if skill.mp_cost > player.mp:
                    print("Not enough MP to use this skill.")
                else:
                    skill.use_skill(player, enemy)
        else:
            print("Invalid skill choice")

        print(f"{player.job_class} Health:", player.hp)
        print("Enemy Health:", enemy.hp)
        
    print("Enemy attacking")
    player.hp -= enemy.ap - player.dp
    print(f"{player.job_class} Health:", player.hp)

    # Check win/lose conditions
    if player.hp <= 0:
        print("Player loses!")
        break
    elif enemy.hp <= 0:
        print("Player wins!")
        break
```

```python
#ajanya's version
import random

class PlayerCharacter:
    def __init__(self, name, hp, max_hp, ap, dp, sp, mp, max_mp, image, job_class, magic_skills):
        self.name = name
        self.hp = hp
        self.max_hp = max_hp
        self.ap = ap
        self.dp = dp
        self.sp = sp
        self.mp = mp
        self.max_mp = max_mp
        self.image = image
        self.job_class = job_class
        self.magic_skills = magic_skills

    def reset_stats(self):
        self.hp = self.max_hp
        self.mp = self.max_mp

class MagicSkill:
    def __init__(self, name, power, skill_type):
        self.name = name
        self.power = power
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, user, target):
        print(f"{user.name} uses {self.name} on {target.name}.")

        if self.skill_type == "heal" and isinstance(target, PlayerCharacter):
            target.hp += self.power
            print(f"{target.name}'s health increased by {self.power}. New HP: {target.hp}")
        elif self.skill_type == "damage" and isinstance(target, PlayerCharacter):
            target.hp -= self.power
            print(f"{target.name} receives {self.power} damage. New HP: {target.hp}")
        else:
            print("Invalid target for this skill.")

# Create magic skills for player 1
heal_skill_p1 = MagicSkill("Healing Light", 20, "heal")
fireball_skill_p1 = MagicSkill("Fireball", 30, "damage")

# Create magic skills for player 2
heal_skill_p2 = MagicSkill("Healing Wave", 25, "heal")
lightning_skill_p2 = MagicSkill("Lightning Strike", 35, "damage")

# Create player characters
player1 = PlayerCharacter(
    name="Player 1",
    hp=100,
    max_hp=100,
    ap=20,
    dp=15,
    sp=10,
    mp=5,
    max_mp=5,
    image="",
    job_class="Warrior",
    magic_skills=[heal_skill_p1, fireball_skill_p1]
)

player2 = PlayerCharacter(
    name="Player 2",
    hp=120,
    max_hp=120,
    ap=18,
    dp=12,
    sp=12,
    mp=8,
    max_mp=8,
    image="",
    job_class="Mage",
    magic_skills=[heal_skill_p2, lightning_skill_p2]
)

# Create enemy characters
enemy1 = PlayerCharacter(
    name="Enemy 1",
    hp=80,
    max_hp=80,
    ap=15,
    dp=10,
    sp=8,
    mp=0,
    max_mp=0,
    image="",
    job_class="Enemy",
    magic_skills=[]
)

enemy2 = PlayerCharacter(
    name="Enemy 2",
    hp=100,
    max_hp=100,
    ap=12,
    dp=8,
    sp=10,
    mp=0,
    max_mp=0,
    image="",
    job_class="Enemy",
    magic_skills=[]
)

# Battle simulation
players = [player1, player2]
enemies = [enemy1, enemy2]

while True:
    for i, player in enumerate(players):
        print(f"{i+1}) {player.name}")
    player_index = int(input("Select a player: ")) - 1

    if 0 <= player_index < len(players):
        player = players[player_index]

        print(f"{player.name} action:")
        print("1) Attack")
        print("2) Use Skill")
        print("3) Defend")
        print("4) Quit")
        action = int(input())

        if action == 1:
            target = random.choice(enemies)
            damage = player.ap - target.dp
            target.hp -= damage
            print(f"{player.name} attacks {target.name} and deals {damage} damage.")
            print(f"{target.name} HP: {target.hp}")
            if target.hp <= 0:
                print(f"{target.name} defeated!")
                enemies.remove(target)
        elif action == 2:
            if player.magic_skills:
                print(f"{player.name} magic skills:")
                for i, skill in enumerate(player.magic_skills):
                    print(f"{i+1}) {skill.name}")
                skill_index = int(input("Choose a skill: ")) - 1
                if 0 <= skill_index < len(player.magic_skills):
                    skill = player.magic_skills[skill_index]
                    target = None
                    if skill.skill_type == "heal":
                        while True:
                            for i, player in enumerate(players):
                                print(f"{i+1}) {player.name}")
                            target_index = int(input("Select a player to heal: ")) - 1
                            if 0 <= target_index < len(players):
                                target = players[target_index]
                                break
                            else:
                                print("Invalid target choice.")
                    elif skill.skill_type == "damage":
                        target = random.choice(enemies)

                    if target is not None:
                        skill.use_skill(player, target)
                else:
                    print("Invalid skill choice.")
            else:
                print(f"{player.name} has no magic skills.")
        elif action == 3:
            player.dp += 5  # Increase player's defense points
            print(f"{player.name} is in a defensive stance. Defense points increased.")
        elif action == 4:
            print(f"{player.name} quits the game.")
            break
    else:
        print("Invalid player choice.")

    if not enemies:
        print("You win!")
        break

    for enemy in enemies:
        target = random.choice(players)
        damage = max(0, enemy.ap - target.dp)
        target.hp -= damage
        print(f"{enemy.name} attacks {target.name} and deals {damage} damage.")
        print(f"2{target.name} HP: {target.hp}")
        if target.hp <= 0:
            print(f"{target.name} defeated!")
            players.remove(target)

    if not players:
        print("You lose!")
        break
```

```python
import random

class HealSkill:
    def __init__(self, name, power, skill_type):
        self.name = name
        self.power = power
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, target, heal_points):
        target.hp += heal_points


class BackstabSkill:
    def __init__(self, name, power, skill_type):
        self.name = name
        self.power = power
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, target, damage):
        ## target.hp -= damage
        target.hp-=self.power
#define two skill for rouge
class ShadowstepSkill:
    def __init__(self, name, power, skill_type):
        self.name = name
        self.power = power
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, target, damage):
        target.hp -= damage
        
class VitatilityBoost:
    def __init__(self, name, power, skill_type):
        self.name = name
        self.power = power
        self.skill_type = skill_type

    def display(self):
        print(f"Skill: {self.name}, Power: {self.power}, Type: {self.skill_type}")

    def use_skill(self, target):
        target.hp += self.power
        print(f"{target.name}'s health increased by {self.power}. New HP: {target.hp}")

                

    
        
class PlayerCharacter:
    def __init__(self, hp, max_hp, ap, dp, sp, mp, max_mp, image, job_class, magic_skills):
        self.hp = hp
        self.max_hp = max_hp
        self.ap = ap
        self.dp = dp
        self.sp = sp
        self.mp = mp
        self.max_mp = max_mp
        self.image = image
        self.job_class = job_class
        self.magic_skills = magic_skills

    def reset_stats(self):
        self.hp = self.max_hp
        self.mp = self.max_mp        
 
        

# Create a character
warrior = PlayerCharacter(
    hp=100,
    max_hp=100,
    ap=20,
    dp=15,
    sp=10,
    mp=5,
    max_mp=5,
    image="A strong and heavily-armed warrior",
    job_class="Warrior",
    magic_skills=[]
)

# Create a second player
rogue = PlayerCharacter(
    hp=100,
    max_hp=100,
    ap=25,
    dp=10,
    sp=15,
    mp=15,
    max_mp=15,
    image="No idea",
    job_class="Rogue",
    magic_skills=[]
)
# add skills specifically
warrior.magic_skills.append(HealSkill("Healing Light", 20, "Heal"))
warrior.magic_skills.append(BackstabSkill("Backstab", 40, "damage"))
rogue.magic_skills.append(ShadowstepSkill("Shadowstep", 35, "damage"))
rogue.magic_skills.append(PoisonedBladeSkill("VitalityBoost", 30, ""))


class EnemyCharacter:
    def __init__(self, hp, ap, dp, sp, image, enemy_type ):
        self.hp=hp
        self.ap=ap
        self.dp=dp
        self.image=image
        self.enemy_type = enemy_type
        
#create enemy characters
goblin = EnemyCharacter(
    hp=100,
    ap=10,
    dp=5,
    sp=6,
    image="A small and dangerous creature",
    enemy_type="goblin"
     )
orc = EnemyCharacter(
    hp=100,
    ap=15,
    dp=4,
    sp=6,
    image="a brutive and aggresive character",
    enemy_type="orc"
     )
while True:
    print("Select a Player Character:")
    print("1) Warrior")
    print("2) Rogue")
    player_choice = int(input("Enter player number: "))

    if player_choice == 1:
        player = warrior
        print("Player Character: Warrior")
    elif player_choice == 2:
        player = rogue
        print("Player Character: rogue")
    else:
        print("Invalid player choice")
        continue

    # Select a random enemy character
    enemy = random.choice([goblin, orc])
    print("Enemy type:", enemy.enemy_type)

    # Display player action menu
    print("Player action:")
    print("1) Attack")
    print("2) Quit")
    print("3) Restart")
    print("4) Use Skill")

    # Get input from player
    action = int(input("Enter action number: "))

    # Process player action
    if player.hp >=enemy.hp:
        if action == 1:
            # Player attacks enemy
            print(f"{player.job_class} attacking")
            enemy.hp -= player.ap
            print("Enemy Health:", enemy.hp)
            if enemy.hp <= 0:
                print("Player wins!")
                break
        elif action == 2:
            print("Player Quits")
            break
        elif action == 3:
            print("Player Restarts")
            player.reset_stats()
            for enemy in [goblin, orc]:
                enemy.hp = enemy.max_hp
            continue
        elif action == 4:
            # Player uses skill
            print(f"Select a Skill for {player.job_class}:")
            for index, skill in enumerate(player.magic_skills, start=1):
                print(f"{index}) {skill.name}")
            skill_choice = int(input("Enter skill number: "))
            if skill_choice >= 1 and skill_choice <= len(player.magic_skills):
                skill = player.magic_skills[skill_choice - 1]
                skill.use_skill(enemy, skill.power)
                print(f"{player.job_class} uses {skill.name} against {enemy.enemy_type}")
            else:
                print("Invalid skill choice")

            print(f"{player.job_class} Health:", player.hp)
            print("Enemy Health:", enemy.hp)
    
        
        # Enemy attacks player
    print("Enemy attacking")
    player.hp -= enemy.ap
    print(f"{player.job_class} Health:", player.hp)
    if player.hp <= 0:
        print("Player loses!")
        break



   
    


```
